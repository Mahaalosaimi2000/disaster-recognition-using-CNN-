```python
# train_img_path = os.path.join(path, "train/images" )
# valid_img_path = os.path.join(path, 'valid/images')
# test_img_path = os.path.join(path,'test/images')
# test_label_path= os.path.join(path, 'test/labels')
# valid_label_path = os.path.join(path,'valid/labels')
# train_label_path = os.path.join(path,'train/labels')

train_img_path = os.path.join(path, "DisasterModel/train" )
valid_img_path = os.path.join(path, 'DisasterModel/validation')
test_img_path = os.path.join(path,'DisasterModel/test')




```

```python
# الكود اخذته من ذا الرابط
#https://medium.com/codex/have-photos-will-model-af804573cfe6
# و
# https://medium.com/data-science-collective/image-augumentation-using-deep-rbm-fc32e2190de8
# و
# https://medium.com/@kumudtraveldiaries/step-by-step-preprocessing-guide-for-images-in-both-cnn-and-dense-layer-pipelines-1994c3ad3e87
# normlize the data from [0,255] to [0,1]
## Initialising the generators for train and test data and validation data
# The rescale parameter ensures the input range in [0, 1
train_datagen = ImageDataGenerator(rescale=1./255,     rotation_range=20,
    zoom_range=0.2)
    # Data augmentation only for the training data
    # We set two types of transformations:
    #rotation_range=20: Randomly rotates images by up to 20 degrees.
    #zoom_range=0.2: Randomly zooms in on images by up to 20%.
test_datagen = ImageDataGenerator(rescale=1./255)
validdatagen = ImageDataGenerator(rescale=1./255)
# Training data
train_generator = train_datagen.flow_from_directory(train_img_path,
                                color_mode='rgb',
                                class_mode='categorical',
                                # reshape each image to add a channel dimension cause CNN expect (H,W,Dim) and dim = 3 if it's RGB image
                                target_size=(224, 224,3),
                                batch_size=32,
                                shuffle=True)
# Validation data
val_generator = validdatagen.flow_from_directory(valid_img_path,
                              color_mode='rgb',
                              class_mode='categorical',
                              target_size=(224, 224,3),
                              batch_size=32,
                              shuffle=False)
test_generator = test_datagen.flow_from_directory(test_img_path,
                              color_mode='rgb',
                              class_mode='categorical',
                              target_size=(224, 224,3),
                              batch_size=32,
                              shuffle=False)
```

```python
# make sure that all of them has the same size and dim =3
print("The target size of train img: ",train_generator.target_size)
print("The target size of valid img: ",val_generator.target_size)
print("The target size of test img: ",test_generator.target_size)
```
